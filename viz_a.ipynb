```
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```
buzz_code = pd.read_csv('CoffeeAndCodeLT2018.csv')
buzz_code.head()
```

```
X = buzz_code['CoffeeSolveBugs'].values
Y = buzz_code['CoffeeCupsPerDay'].values
```

```
plt.figure()
x1 = X
x2 = Y
plt.plot(x1, label='Coffee Can Solve Bugs?', color='green')
plt.plot(x2, label='Cups of Coffee per Day', color='pink')
plt.legend()
plt.show()
```
